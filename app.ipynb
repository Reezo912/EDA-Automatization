```python
# Compruebo si hay filas duplicadas
'''print(f'Hay {df.duplicated().sum()} filas duplicadas')

# Comprobamos si hay PlayerID o Name repetidos.

print(f'Hay {df["PlayerID"].duplicated().sum()} playerID duplicados')
print(f'Hay {df["Name"].duplicated().sum()} Name duplicados')'''



def comprobar_duplicados(dataframe):

    print(dataframe.head(10))

    keep_looking = True

    while keep_looking:
        
        user_input: str = input('Que columna deseas comprobar duplicados, escribe "no" si quieres salir: ')
        
        if user_input.lower() == 'no':
            keep_looking = False
          
        if user_input not in dataframe.columns:
            print(f"La columna '{user_input}' no existe. Las columnas disponibles son: {list(dataframe.columns)}")

        else:
            duplicate_count = dataframe[user_input].duplicated().sum()
            print(f'Hay {duplicate_count} duplicados en la columna {user_input}')
            
        print(dataframe.columns)

comprobar_duplicados(df)
```

```python
def borrar_columnas(dataframe) -> None:
    
    print(dataframe.head(10))
    keep_deleting = True

    while keep_deleting:
        
        user_input: str = input('Que columna deseas borrar, escribe "no" si quieres salir: ')
        
        if user_input.lower() == 'no':
            keep_deleting = False
          
        else:
            dataframe.drop(user_input, axis=1, inplace=True)\
            
        print(dataframe.columns)


borrar_columnas(df)
```
